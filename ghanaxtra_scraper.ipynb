```python
url_list = []
for i in range(1,len(parse_urls)):
    orig_url = parse_urls[i][2]
    tstamp = parse_urls[i][1]
    waylink = tstamp+'/'+orig_url
    url_list.append(waylink)
```

```python
final_url = []

for url in url_list:
    final_url.append("http://web.archive.org/web/"+url)
```

```python
w_links = {}
w_list=[]
r=rq.get(final_url[1]).text
soup=bs(r,'html.parser')
a_links = soup.find_all('a')
for link in a_links:
    processed_link=re.findall(r'<a\s+.*?href="(.*?\.html)".*?>', str(link))
    for i in processed_link:
        w_list.append(i)

final_links = list(set(w_list))        
df = pd.DataFrame(final_links)
df.to_csv('example.csv')
```

```python

data_list = []

for url in data['links'].values:
    post_paragraphs = [] #reset the post_paragraph list to avoid joining posts from different urls
    r = rq.get(url)
    s = bs(r.content, 'html.parser')
    try:
        p_tags = s.find_all('p')
        for p_tag in p_tags:
            post = p_tag.text
            post_paragraphs.append(post)
        
        p_sentence = ''.join(post_paragraphs)
        
        post_title = s.find('h1').text
        author = s.find('div', class_="td-post-author-name").find('a').text
    except:
        post = None
        post_title = None
        author = None
        
    content = {
        'post_link': url,
        'post': p_sentence,
        'title': post_title,
        'author': author
    }
    
    data_list.append(content)
    
    
with open('data.json', 'w') as file:
    json.dump(data_list, file)
```
